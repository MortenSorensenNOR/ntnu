```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
u = np.linspace(0, 1, 100)
v = np.linspace(0, 1, 100)
U, V = np.meshgrid(u, v)
X = np.vstack([U.ravel(), V.ravel()])
```

```python
def g(x):
    return np.dot(np.array([[2, 1], [1, 2], [1, 1]]), x)
```

```python
G = g(X)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# Plot the surface
ax.plot_surface(G[0, :].reshape(U.shape), G[1, :].reshape(U.shape), G[2, :].reshape(U.shape), cmap='viridis')

# Set an equal aspect ratio
ax.set_aspect('equal')

plt.show()
```

```python
u = np.linspace(0, 2*np.pi, 100)
v = np.linspace(0, np.pi, 100)
U, V = np.meshgrid(u, v)
X = np.vstack([U.ravel(), V.ravel()])
```

```python
def g(x):
    return np.array([np.cos(x[0]) * np.sin(x[1]), np.sin(x[0]) * np.sin(x[1]), np.cos(x[1])])

G = g(X)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# Plot the surface
ax.plot_surface(G[0, :].reshape(U.shape), G[1, :].reshape(U.shape), G[2, :].reshape(U.shape), cmap='viridis')

# Set an equal aspect ratio
ax.set_aspect('equal')

plt.show()
```

```python
u = np.linspace(0, 2*np.pi, 100)
v = np.linspace(0, 1, 100)
U, V = np.meshgrid(u, v)
X = np.vstack([U.ravel(), V.ravel()])
```

```python
def p(z):
    return (40 * z - 4) * (z - 1)**4

def g(x):
    return np.array([np.cos(x[0]) * p(x[1]), np.sin(x[0]) * p(x[1]), x[1]])

G = g(X)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# Plot the surface
ax.plot_surface(G[0, :].reshape(U.shape), G[1, :].reshape(U.shape), G[2, :].reshape(U.shape), cmap='viridis')

ax.view_init(elev=0, azim=0)  # Set elevation and azimuth

plt.show()
```
